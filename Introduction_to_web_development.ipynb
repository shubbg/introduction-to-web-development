#Introduction to web development

1.Explain the main components of client-server architecture and their functions

  ->1. Client
 What it is: The user's device (like a phone, computer, or app).

 Function: It sends requests to the server to get data or perform some action.

 Example: When you open YouTube and click on a video, your browser (client) sends a request to the YouTube server.

 2.Server
 What it is: A powerful computer that stores data and programs.

 Function: It receives requests from the client, processes them, and sends back a response.

 Example: YouTube’s server sends you the video and comments when you request them.

 3.Network (Internet)
 What it is: The connection between client and server (usually the internet).

 Function: It helps send and receive messages (requests and responses) between the client and server.

 Example: Your Wi-Fi or mobile data lets your device talk to the server.

 4.Database (on the server)
 What it is: A place where data is stored and organized.

 Function: It saves data like user accounts, messages, files, etc., and allows the server to fetch or update them when needed.

 Example: When you log in, the server checks your username/password in the database.

 4.Application (Software)
 What it is: The code or program running on the server.

 Function: It processes the client’s request, does the logic, and prepares a response.

 Example: If you search “cat videos,” the server software finds matching videos and sends the list to your screen.

2.Discuss the various career paths available after completing a web development course

  ->1. Front-End Developer
Builds the visual part of websites using HTML, CSS, and JavaScript. Focuses on design, layout, and user experience.

2. Back-End Developer
Works on server-side logic, databases, and APIs using languages like Node.js, PHP, Python, or Java.

3. Full-Stack Developer
Handles both front-end and back-end development. Can build a complete website or web application independently.

4. UI/UX Designer
Focuses on designing user-friendly and attractive interfaces. Works with design tools like Figma or Adobe XD.

5. E-commerce Developer
Builds and manages online stores using platforms like Shopify, WooCommerce, or Magento.

6. Web Application Developer
Creates interactive and dynamic web apps using frameworks like React, Angular, or Vue.js.

7. DevOps Engineer (with extra training)
Manages hosting, deployment, and automation using tools like Docker, Jenkins, and cloud services like AWS or Azure.

8. Freelancer / Web Consultant
Works independently on various web projects for clients, offering flexibility and the potential to earn well.

9. Startup Founder / Tech Entrepreneur
Uses web development skills to build and launch a product or service as a business.

4.What are the advantages of using a Version Control System like Git in a development project

  ->1. Tracks Changes
Git keeps a record of every change made to your code, so you can go back to an earlier version if something breaks.

2. Team Collaboration
Multiple people can work on the same project at the same time without messing up each other’s work.

3. Backup
Your code is stored safely. Even if your computer crashes, you won’t lose your project (especially when using GitHub).

4. Branching
You can create separate branches to try out new features or fix bugs without affecting the main project.

5. Easy to Merge
You can combine different versions of code easily, especially when working in teams.

6. See Who Did What
Git shows who made each change and when. This helps in understanding and managing the project better.

7. Better Project Management
Git helps organize your project, keep it clean, and track progress over time.


5.Compare and contrast a text editor and an IDE, highlighting their key features and uses

  ->1. Text Editor
What it is:
A lightweight tool used to write and edit code.

Key Features:

Simple and fast

Syntax highlighting (color-coded code)

Line numbers

Basic file navigation

Examples: Notepad++, Sublime Text, VS Code (can act as both)

Use:
Good for small projects, quick edits, or learning to code.

2. IDE (Integrated Development Environment)
What it is:
A complete software package for building, running, and debugging code.

Key Features:

Everything in a text editor plus:

Code auto-completion

Built-in compiler or interpreter

Debugging tools

Project management

Version control tools

Examples: IntelliJ IDEA, Eclipse, PyCharm, Visual Studio

Use:
Best for large projects and professional development where you need many tools in one place.
